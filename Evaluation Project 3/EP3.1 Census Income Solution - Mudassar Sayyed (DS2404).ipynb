### Treating Outliers using Z-Score method

```python
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(df))
# Define threshold for outliers
threshold=3
np.where(z>3)
```

```python
df1=df[(z<3).all(axis=1)]
df1
```

```python
#Using IQR method to remove Outliers 
from scipy import stats
IQR = stats.iqr(df)
IQR
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
```

```python
IQR = df[~ ((df < (Q1 -1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(IQR.shape)
```

#### Using the IQR method, we are left with only 16,289 rows, while applying the Z-score method retains 29,428 rows. Therefore, we will use the Z-score as our outlier elimination technique.
